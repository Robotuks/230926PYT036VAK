## Class

```python
class Book:
    def __init__(self, title, author, book_id, avaliable=True):
        self.title = title
        self.author = author
        self.book_id = book_id
        self.avaliable = avaliable

    def __str__(self):
        return f"{self.title} by {self.author}"


class Library():
    def __init__(self):
        self.books = []
        self.gone_books = []

    def add_book(self, book):
        self.books.append(book)
    
    def all_books(self):
        for book in self.books + self.gone_books:
            print(book)

    def available_books(self):
        for book in self.books:
            print(book)

    def taken_books(self):
        for book in self.gone_books:
            print(book)

    def take_book(self, book_id):
        for book in self.books:
            if book_id == book.book_id and book.avaliable == True:
                self.books.remove(book)
                self.gone_books.append(book)
                book.avaliable = False
                return f"\nBook taken: {book}"
        return "\nBook is not avaliable."

    def return_book(self, book_id):
        for book in self.gone_books:
            if book_id == book.book_id:
                self.books.append(book)
                self.gone_books.remove(book)
                book.avaliable = True
                return f"\n{book} was returned"
        return "\nCan't return."


book_1 = Book("The Book", "O.K. Author", 1)
book_2 = Book("The Best Book", "Cool A. Author", 2)
book_3 = Book("The Worst Book", "Bad J. Author", 3)
book_4 = Book("The Meh Book", "Decent K. Author", 4)
book_5 = Book("The Meh Book", "Decent K. Author", 5)

library = Library()
library.add_book(book_1)
library.add_book(book_2)
library.add_book(book_3)
library.add_book(book_4)
library.add_book(book_5)

print(book_1.__str__())

print("\n   Books in the library: \n")
library.all_books()

print(library.take_book(1))
print(library.take_book(3))
print(library.take_book(4))


print("\n   Books in the library: \n")
library.all_books()

print("\n   Available books: \n")
library.available_books()

print("\n   Taken books: \n")
library.taken_books()


print(library.return_book(1))

print("\n   Books in the library: \n")
library.all_books()

print("\n   Taken books: \n")
library.taken_books()
```

```python
class ATM:
    def __init__(self, cash_in_atm, bank, availability):
        self.cash_in_atm = cash_in_atm
        self.bank = bank
        self.availability = availability
        
    def is_working(self):
        if self.availability:
            print(f'Welcome to {self.bank}\'s ATM - please insert your card')
        else:
            print('Apologies for technical issues')
        return self.availability
            
    def withdraw_money(self,requested_amount):
        if self.is_working():
            if requested_amount < self.cash_in_atm:
                self.cash_in_atm = self.cash_in_atm - requested_amount
                print('Here your money')
            else:
                print('Sorry, not enough cash')

            
    def deposit(self,deposit_money):
        if self.is_working():
            self.cash_in_atm = self.cash_in_atm + deposit_money
            print(f'{deposit_money} were added to your account')

# TODO - create Client class

swed = ATM(10000, 'Swedbank', True)
seb = ATM(12000, 'Seb', False)
luminor = ATM(2000, 'Luminor', True)

#input values:
requested_amount = int(input('Please enter the amount you want to borrow'))
deposit_money = int(input('Please enter the amount you want to deposit'))

#calling the functions:
swed.is_working()
swed.withdraw_money(requested_amount)
swed.deposit(deposit_money)

seb.is_working()
seb.withdraw_money(requested_amount)
seb.deposit(deposit_money)

luminor.is_working()
luminor.withdraw_money(requested_amount)
luminor.deposit(deposit_money)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,100)
print(x)
print(type(x) )
y = x * 2
print(y)
z = x ** 2
print(z)
```

```python
fig = plt.figure()
ax = fig.add_axes( [0,0,1,1] )

ax.plot(x, y)

ax.set_title('My title')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
```

```python
fig = plt.figure()
ax = fig.add_axes( [0,0,1,1] )

z[60] = 4000

ax.plot(x, z)

ax.set_title('My title')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')

ax_2 = fig.add_axes( [0.4, 0.5, 0.2, 0.2] ) # [left, bottom, width, height] 
ax_2.plot(x, z)
ax_2.set_xlim(55,65)
ax_2.set_ylim(3000,4500)
ax_2.set_title('Zoomed in')
```

```python
fig, axes = plt.subplots(1, 2)
print(axes.shape)
axes[0].plot(x, y, color='blue', lw=5, ls='--')
axes[1].plot(x, z, color='red', lw=3)
```

```python
fig, axes = plt.subplots(2,4)
print(axes.shape)
for row in axes:
    # print (len(row))
    for ax in row:
        ax.plot(x,z)
```
