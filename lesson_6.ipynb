## local vs global scope

```python
some_global_variable = "GLOBAL"

# def substract_all( *args): # args = arguments
def substract_all(that_global_val, *args): # args = arguments
    """
    that_global_val - constant
    """
    print(args)
    print(some_global_variable) # BAD practice. Don't do this
    print(that_global_val) # create argument
    function_result = args[0]
    for num in args[1:]:
        function_result -= num
    
    return function_result
```

```python
substract_all(1,2)
# print(function_result) # error. Can't access function local varialbe
print()
substract_all(some_global_variable, 1,2,3)
```

```python
some_global_variable = "GLOBAL"

# def substract_all( *args): # args = arguments
def substract_all(*args): # args = arguments
    """
    that_global_val - constant
    """
    print(args)
    # DON'T use global. Bad practice
    global some_global_variable # say you want to overwrite global variable
    some_global_variable = 10
    print(some_global_variable) # BAD practice. Don't do this

    function_result = args[0]
    for num in args[1:]:
        function_result -= num
    
    return function_result
```

```python
substract_all(-10,1,1)
print(some_global_variable)
```

## kwargs = key arguments

```python
def kwargs_f( **kwargs ):
    print(kwargs)
```

```python
kwargs_f()
kwargs_f(name='Darius', number=10)
kwargs_f(name='Darius', number=10, is_teacher=True)
```

```python
def kwargs_f( **kwargs ):
    # print(kwargs)
    for k,v in kwargs.items():
        print(f'key "{k}" has a value of {v}')
```

```python
kwargs_f()
kwargs_f(name='Darius', number=10)
```

```python
def say_hello_kwargs(name, surname, **kwargs):
    print(f'Hello {name} {surname}')

    if 'season' in kwargs.keys():
        print(f"{ kwargs['season'] } is so beutiful")
        kwargs.pop('season')
```

```python
# say_hello_kwargs() # error
say_hello_kwargs('Darius', 'Grigaliunas')
say_hello_kwargs('Darius', 'Grigaliunas', season='summer')
```
